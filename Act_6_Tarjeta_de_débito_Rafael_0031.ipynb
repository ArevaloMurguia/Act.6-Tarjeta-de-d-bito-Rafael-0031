### Clase: Tarjeta de débito

```python
print("===== CONSULTAR SALDO =====")

saldo = 5000
nip = 1234

clave = int(input("Ingresa NIP: "))

if clave == nip:
    print("Saldo actual:", saldo)

else:
    print("NIP incorrecto")
```

```python
print("===== RETIRAR DINERO =====")

saldo = 5000
nip = 1234

clave = int(input("Ingresa NIP: "))

if clave == nip:

    retiro = int(input("Cantidad a retirar: "))

    if retiro <= saldo:

        saldo = saldo - retiro

        print("Retiro realizado")
        print("Nuevo saldo:", saldo)

    else:
        print("Saldo insuficiente")

else:
    print("NIP incorrecto")
```

```python
print("===== DEPOSITAR DINERO =====")

saldo = 5000
nip = 1234

clave = int(input("Ingresa NIP: "))

if clave == nip:

    deposito = int(input("Cantidad a depositar: "))

    saldo = saldo + deposito

    print("Depósito realizado")
    print("Nuevo saldo:", saldo)

else:
    print("NIP incorrecto")
```
