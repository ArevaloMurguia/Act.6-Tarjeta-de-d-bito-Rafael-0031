```python
# ==================================================
# Rafael Emiliano Arevalo Murguia
# No. Lista: 3
# Clase: Tarjeta de Debito
# IA Simulada con if, elif, else y while
# ==================================================

print("======================================")
print(" ASISTENTE IA DE TARJETA DE DEBITO ")
print("======================================")

continuar = "si"

while continuar == "si":

    print("\nHola, soy tu asistente financiero.")
    print("Te ayudare a analizar el uso de tu tarjeta de debito.\n")

    nombre = input("¿Cual es tu nombre?: ")

    print("\nMucho gusto,", nombre)

    saldo = float(input("\nIngresa tu saldo disponible: "))
    retiro = float(input("Ingresa el monto que deseas retirar: "))
    compras = int(input("¿Cuantas compras realizaste hoy?: "))

    print("\nAnalizando informacion...")

    if saldo <= 0:
        print("El saldo ingresado no es valido.")

    elif retiro < 0:
        print("El retiro ingresado no es valido.")

    elif compras < 0:
        print("La cantidad de compras no es valida.")

    else:

        saldo_restante = saldo - retiro

        print("\n=========== RESULTADOS ===========")
        print("Saldo inicial:", saldo)
        print("Retiro realizado:", retiro)
        print("Compras realizadas:", compras)
        print("Saldo restante:", saldo_restante)

        print("\n=========== ANALISIS IA ===========")

        if saldo_restante < 0:
            print("ALERTA: No cuentas con fondos suficientes.")

        elif saldo_restante <= 1000:
            print("La IA detecta saldo bajo.")

        else:
            print("La IA detecta saldo saludable.")

        if compras <= 3:
            print("Uso moderado de la tarjeta.")

        elif compras > 3 and compras <= 10:
            print("Uso frecuente de la tarjeta.")

        else:
            print("Uso intensivo de la tarjeta.")

        if retiro > saldo * 0.8:
            print("La IA detecta un retiro elevado.")

        else:
            print("El retiro parece normal.")

        print("\n=========== RECOMENDACION ===========")

        if saldo_restante < 500:
            print("Se recomienda evitar gastos innecesarios.")

        elif saldo_restante >= 500 and saldo_restante <= 5000:
            print("Tus finanzas parecen estables.")

        else:
            print("Excelente manejo de tu dinero.")

    continuar = input("\n¿Deseas realizar otro analisis? (si/no): ")

print("\nGracias por utilizar el asistente IA.")
print("Programa finalizado.")
```
